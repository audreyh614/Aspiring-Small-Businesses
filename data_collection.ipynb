In this notebook you will observe the code I used to conduct a webscrape on California businesses, ranging from small to medium, that were approved for loans through the Payment Protection Program (PPP).


```python
import requests

import pandas as pd

from sqlalchemy import create_engine

from bs4 import BeautifulSoup
```

```python
url = 'https://ppp.pubb.org/ca'
```

```python
params = {'q':'*'}
```

```python
web_request = requests.get(url, params=params)
```

```python
web_request
```

```python
web_request.text
```

```python
soup = BeautifulSoup(web_request.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
soup.findAll('tr')
```

```python
company_info = soup.findAll('tr')
```

```python
company_info
```

```python
company_details = {
    'business_name': [],
    'address': [],
    'loan_amount': [],
    'link': []
} 
for company in company_info:
    
    if company.find('a') is None:
        continue
        
    business_name = company.find('a').text
    company_details['business_name'].append(business_name)
    if business_name is not None:
        print('Business Name:', business_name)
    
    address = company.findAll('td')[1].text
    company_details['address'].append(address)
    if address is not None:
        print('Address:', address)
    
    loan_amount = company.findAll('td')[2].text
    company_details['loan_amount'].append(loan_amount)
    if loan_amount is not None:
        print('Loan Amount:', loan_amount)
             
    link = company.find('a')['href']
    company_details['link'].append(link)
    if link is not None:
        print('Link:', 'https://ppp.pubb.org'+link)
    
    print('-'*115)
```

```python
company_details
```

```python
loans_df = pd.DataFrame(company_details)
```

```python
## DO NOT RUN THIS CELL! Unless you want to change your CaliforniaLoanApplication table data
for PageNumber in range(1,5):
    print('Page:', PageNumber)
    
##you stopped at page 5! which returned 250 rows/ businesses: url = 'https://ppp.pubb.org/ca/?page=5'
    url = 'https://ppp.pubb.org/ca/?page=' 

    params = {'q':'*', 'state':'California','pn': PageNumber}

    web_request = requests.get(url, params=params)

    soup = BeautifulSoup(web_request.text, 'html.parser')

    company_info = soup.findAll('tr')
    

    company_details = {
        'business_name': [],
        'address': [],
        'loan_amount': [],
        'link': []
    } 
    for company in company_info:

        if company.find('a') is None:
            continue

        business_name = company.find('a').text
        company_details['business_name'].append(business_name)
        if business_name is not None:
            print('Business Name:', business_name)

        address = company.findAll('td')[1].text
        company_details['address'].append(address)
        if address is not None:
            print('Address:', address)

        loan_amount = company.findAll('td')[2].text
        company_details['loan_amount'].append(loan_amount)
        if loan_amount is not None:
            print('Loan Amount:', loan_amount)

        link = company.find('a')['href']
        company_details['link'].append(link)
        if link is not None:
            print('Link:', 'https://ppp.pubb.org'+link)

        print('-'*115)
    
    loans_df = pd.DataFrame(company_details)
    
    loans_df.to_sql('CaliforniaLoanApplication', engine, if_exists="append", index=False)
```
